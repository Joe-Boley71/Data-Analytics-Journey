### Create the table AUTHORS

```python
%%sql
CREATE TABLE authors (
    author_id  INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name         NOT NULL
                       CHECK (length(trim(first_name) ) > 0),
    last_name  TEXT    NOT NULL
                       CHECK (length(trim(last_name) ) > 0),
    biography  TEXT
);
```

### Create the table BOOKS

```python
%%sql
CREATE TABLE books (
    book_id          INTEGER PRIMARY KEY AUTOINCREMENT,
    title            TEXT    NOT NULL
                             CHECK(length(trim(title)) > 0),
    isbn             TEXT    NOT NULL
                             UNIQUE
                             CHECK(length >= 10),
    publication_year INTEGER CHECK(publication_year >= 1455 AND
                                    publication_year <= strftime('%Y', 'now') ),
    price            REAL    NOT NULL
                             CHECK(price > 0.0),
    stock_quantity   INTEGER NOT NULL
                             DEFAULT (0) 
                             CHECK(stock_quantity > 0),
    author_id        INTEGER NOT NULL
                             REFERENCES authors (author_id) 
);
```

### Create the table CUSTOMERS

```python
%%sql
CREATE TABLE customers (
    customer_id      INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name       TEXT    NOT NULL
                             CHECK (length(trim(first_name) ) > 0),
    last_name        TEXT    CHECK (length(trim(last_name) ) > 0) 
                             NOT NULL,
    email            TEXT    UNIQUE
                             NOT NULL
                             CHECK (email LIKE '%_@__%._%'),
    phone            TEXT    CHECK (phone IS NULL OR
                                    length(phone) >= 7),
    shipping_address TEXT    NOT NULL
                             CHECK (length(trim(shipping_address) ) > 0) 
);
```

### Create the table ORDERS

```python
%%sql
CREATE TABLE orders (
    order_id     INTEGER PRIMARY KEY AUTOINCREMENT,
    order_date   TEXT    NOT NULL
                         DEFAULT (CURRENT_TIMESTAMP),
    total_amount REAL    NOT NULL
                         CHECK (total_amount >= 0.0),
    quantity             NOT NULL
                         DEFAULT (1) 
                         CHECK (quantity >= 1),
    customer_id  INTEGER NOT NULL
                         REFERENCES customers (customer_id) ON DELETE RESTRICT,
    book_id      INTEGER NOT NULL
                         REFERENCES books (book_id) ON DELETE RESTRICT
);
```
