```python
# Part 2: Data Cleaning

# 1. Handling Missing Values

# Create copies of original DataFrames for cleaning

df_customers_new = df_customers.copy()
df_inventory_new = df_inventory.copy()
df_products_new = df_products.copy()
df_sales_new = df_sales.copy()
df_stores_new = df_stores.copy()
```

```python
# Verifying the copies of the dataframes existing

df_customers_new
df_inventory_new
df_products_new
df_sales_new
df_stores_new
```

```python
# Verifying which columns have missing values

# df_customers_new.isna()
# df_inventory_new.isna()
# df_products_new.isna()
# df_sales_new.isna()
# df_stores_new.isna()

# For numerical columns with missing values, replace with column mean

# For categorical columns with missing values, replace with most frequent value
```

```python
# for the Customers Table

'''
Logic:
using the .to_datetime function to change the date values to datetime for analysis

recognizing that email, phone, and zip_code are categorical, using the mode values. With Python indexing at 0, Choosing specifically
the pull only the first one, or [0]
'''

df_customers_new['registration_date'] = pd.to_datetime(df_customers_new['registration_date'])

df_customers_new['email'] = df_customers_new['email'].fillna(df_customers_new['email'].mode()[0])
df_customers_new['phone'] = df_customers_new['phone'].fillna(df_customers_new['phone'].mode()[0])
df_customers_new['zip_code'] = df_customers_new['zip_code'].fillna(df_customers_new['zip_code'].mode()[0])
df_customers_new['registration_date'] = df_customers_new['registration_date'].fillna(df_customers_new['registration_date'].median())
```

```python
# For the Inventory Table

'''
logic:
using .to_datetime to change date values as previously mentioned
'''

df_inventory_new['last_restock_date'] = pd.to_datetime(df_inventory_new['last_restock_date'])

df_inventory_new['last_restock_date'] = df_inventory_new['last_restock_date'].fillna(df_inventory_new['last_restock_date'].median())
df_inventory_new['reorder_level'] = df_inventory_new['reorder_level'].fillna(df_inventory_new['reorder_level'].mean())
```

```python
# For the Products Table

'''
logic:

Using .mean as "weight" is a numerical value
'''

df_products_new['weight'] = df_products_new['weight'].fillna(df_products_new['weight'].mean())
```

```python
# For the Sales Table

'''
logics:

Using .to_datetime for help later

'''

df_sales_new['date'] = pd.to_datetime(df_sales_new['date'])

df_sales_new['date'] = df_sales_new['date'].fillna(df_sales_new['date'].median())
df_sales_new['store_id'] = df_sales_new['store_id'].fillna(df_sales_new['store_id'].mode()[0])
df_sales_new['payment_method'] = df_sales_new['payment_method'].fillna(df_sales_new['payment_method'].mode()[0])
```

```python
# For the Stores Table, size_sqft	opening_date

'''

logics:

Using .to_datetime for analysis purposes later
Outside of, recognizing that dates and squarefeets are numerical values that can be manipulated, so using .mean for those

'''

df_stores_new['opening_date'] = pd.to_datetime(df_stores_new['opening_date'])

df_stores_new['opening_date'] = df_stores_new['opening_date'].fillna(df_stores_new['opening_date'].median())
df_stores_new['size_sqft'] = df_stores_new['size_sqft'].fillna(df_stores_new['size_sqft'].mean())
```

### Answer 1

Being able to combine and chain commands such as isna() with any() lets you figure out easily if any columns have any missing values. After, you can chain isna() with sum() to figure out how much missing values exist. ### Answer 2

*How would your approach to handling missing values differ if the missing data was not random but had a pattern or meaning?*

The approach to handling missing values would differ if the missing data was pattern/meaning in how I would go about fixing it. If it was random, then I would use the very same instructions used in this workshop, but if there's a pattern to it, then I would take extra time to investigate the lead from the pattern or meaning, and then approach fixing it from the info I got from that lead. ### Answer 3

*What types of data quality issues might not be immediately visible through simple DataFrame inspection methods?*

Certain data quality issues that might not be visible is how numeric values might become strings, which will cause issues with having to convert them to numerical datatypes for observations. Other issues are akin to values that a human would consider null, but python wouldn't. For example, the values of *"NA", "N/A", "Null"*, etc will be considered as non-nulls to the interpreter when they would be considered as null values to the average human. One other data quality issue that will arise from using only the inspection methods is logic issues that won't be caught by the computer, which is an extension of logic errors in other programming languages. The interpreter and compilers won't catch these, it's up to the human to catch them. Some of these issues are dates being formatted differently as well as trailing spaces which would make **John Doe** different from **John Doe ** (with a trailing space) in the eyes of a programming language.

### Answer 5

*In what scenarios might it be better to remove rows with missing values rather than imputing them?*

The situations that will be useful to remove rows with missing values rather than imputing with various values will be the situations where you'll need upmost precision with the datasets. Healthcare companies will not tolerate guesses when it comes to various aspects such as medicine dosages. Another situation that will be useful to remove rows with missing values rather than imputing values is if that values missing are analytically insignificant. An email value associated with patrons won't be considered **analytically significant** for analysis as the analysis will focus more on the sales of books and the relations between the patrons and books being sold or rented out.
